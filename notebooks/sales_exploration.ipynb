# Nirvana Retail Intelligence Research
This notebook is for ad-hoc financial research, trend visualization, and data exploration.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from intelligence.data_loader import load_sales, load_inventory

# Set style
sns.set_theme(style="whitegrid")
plt.rcParams['figure.figsize'] = [12, 6]
```

## 1. Load Data

```python
sales = load_sales()
print(f"Loaded {len(sales)} sales records")
sales.head()
```

## 2. Sales Trend Visualization

```python
sales['day'] = sales['date'].dt.floor('D')
daily_sales = sales.groupby('day')['total_with_tax'].sum().reset_index()
sns.lineplot(data=daily_sales, x='day', y='total_with_tax')
plt.title('Daily Total Sales (All Shops)')
plt.show()
```
